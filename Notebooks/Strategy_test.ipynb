```python
import matplotlib.pyplot as plt
import numpy as np

df['EURUSD_signal'] = np.where(df['EURUSD=X_Momentum'] > 0, 1, -1)
fig, ax1 = plt.subplots(figsize=(30, 6))
ax1.set_facecolor("black")

# Primary y-axis: EURUSD price
ax1.plot(df["Date."], df["EURUSD=X.Close"], label="EURUSD Price", color="deepskyblue")
ax1.scatter(df["Date."][df['EURUSD_signal'] == 1], df["EURUSD=X.Close"][df['EURUSD_signal'] == 1],
            color="lime", label="Long Signal", marker="^", alpha=0.7)
ax1.scatter(df["Date."][df['EURUSD_signal'] == -1], df["EURUSD=X.Close"][df['EURUSD_signal'] == -1],
            color="red", label="Short Signal", marker="v", alpha=0.7)
ax1.set_ylabel("EUR/USD Price")

# Secondary y-axis: Momentum indicator
ax2 = ax1.twinx()
ax2.plot(df["Date."], df["EURUSD=X_Momentum"], color="orange", label="Momentum")
ax2.set_ylabel("Momentum")

# Labels and title
fig.suptitle("Momentum Strategy: EURUSD", color="white")
ax1.set_xlabel("Date")

# Combine legends from both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc="best")

plt.title('Momentum Strategy: EURUSD')
plt.show()



#this code something is not right even though the code is passing and giving what i need
```

Test for Mean Reversion Strategy

```python
from Strategy.mean_reversion import BollingerMeanReversionStrategy
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# create dummy price data
np.random.seed(42)
data = pd.DataFrame({
    'EURUSD=X.Close': np.random.normal(1.10, 0.01, 100)  # around 1.10 USD  0.01 noise
})

# instantiate and generate signals
bb_test = BollingerMeanReversionStrategy(window=10, num_std=2)
df_test = bb_test.generate_signals(data, price_col='EURUSD=X.Close')

# visualize to understand if signals align with logic
plt.figure(figsize=(30,6))
plt.plot(df_test['EURUSD=X.Close'], label='Price')
plt.plot(df_test['rolling_mean'], label='Rolling Mean')
plt.plot(df_test['upper_band'], label='Upper Band', linestyle='--')
plt.plot(df_test['lower_band'], label='Lower Band', linestyle='--')

# highlight buy/sell points
plt.scatter(df_test.index[df_test['signals'] == 1], df_test['EURUSD=X.Close'][df_test['signals'] == 1],
            label='Buy Signal', marker='^', color='green')
plt.scatter(df_test.index[df_test['signals'] == -1], df_test['EURUSD=X.Close'][df_test['signals'] == -1],
            label='Sell Signal', marker='v', color='red')

plt.title('Bollinger Bands Mean Reversion Strategy  EUR/USD')
plt.xlabel('Date')
plt.ylabel('Price (EUR/USD)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
